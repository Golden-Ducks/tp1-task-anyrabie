```python
# Documents
D1 = "I love cats"
D2 = "Cats are chill"
D3 = "I am late"

# Your Task: implement context window vectorization
# with window size = 1 (so each window is 3 tokens wide)
# Use <s> and </s> padding flags
```

```python
def add_padding(tokens):
    # wrap tokens with start and end flags
    tokens.insert(0, "<s>")
    tokens.insert(len(tokens), "</s>")
    return tokens
```

```python
def extract_windows(tokens, win_size):
    windows = []

    for i in range(win_size, len(tokens) - win_size):
        window = tokens[i - win_size : i + win_size + 1]
        windows.append(window)

    return windows


tokens = ['rabie', 'is', 'boss', 'of', 'hakim']
print(extract_windows(tokens, 1))
```

```python
def build_vocab(all_windows):
    # collect unique windows using a set for efficiency
    unique_windows = set(all_windows)
    # convert to list and sort alphabetically to ensure consistent indexing
    unique_windows = sorted(list(unique_windows))
    # assign index
    vocab = {window: i for i, window in enumerate(unique_windows)}
    return vocab
```

```python
def vectorize_doc(doc_windows, vocab):
    # return a binary vector: 1 if window in doc, 0 otherwise
    return [1 if window in doc_windows else 0 for window in vocab]

# binary_vec = vectorize_doc(all_win, vocab)
# print(binary_vec)
```

```python
# Run (this cell contain additionnal code to fixing some errors)
all_docs = [D1, D2, D3]
all_docs = [doc.lower() for doc in all_docs]
all_tokens = [doc.split() for doc in all_docs]
all_tokens = [add_padding(tokens) for tokens in all_tokens]

# Flatten the list of lists of windows into a single list of windows (tuples)
all_windows_flat = []
for tokens in all_tokens:
    # extract_windows returns a list of lists. Convert each list window to a tuple.
    extracted_windows_for_doc = extract_windows(tokens, win_size=1)
    all_windows_flat.extend([tuple(window) for window in extracted_windows_for_doc])

# Convert to a tuple of all unique windows for build_vocab
# This ensures that 'all_windows' is a flat sequence of hashable tuples
print(all_windows_flat)
all_windows = tuple(all_windows_flat) # Now, this will be a tuple of tuples, which is hashable.

vocab = build_vocab(all_windows)
print("Vocabulary:", vocab)

# Vectorize each document
vectorized_docs = []
for tokens in all_tokens:
    doc_windows_list = extract_windows(tokens, win_size=1)
    # Convert each window in doc_windows to a tuple for consistency with vocab
    doc_windows_tuples = [tuple(w) for w in doc_windows_list]
    vectorized_docs.append(vectorize_doc(doc_windows_tuples, vocab))

print("Binary Vectors for all documents (binary_vec):")
for i, vec in enumerate(vectorized_docs):
    print(f"Document {i+1} vector: {vec}")

print("\nAll tokens (for reference):")
print(all_tokens)
```
