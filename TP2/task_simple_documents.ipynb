```python
D1 = "I feel good."
D2 = "I do feel even better now, thank you!"
D3 = "What's happened to you? you all good?"
```

```python
import string
import spacy

def preprocessing_document(D):
    D = D.lower()
    #Remove punctuation
    for p in string.punctuation:
        if p in D:
            D = D.replace(p,'')

    #Fix contraction
    common_contractions = {
        "can't": "cannot",
        "won't": "will not",
        "i'm": "i am",
        "he's": "he is",
        "she's": "she is",
        "it's": "it is",
        "we're": "we are",
        "they're": "they are",
        "i've": "i have",
        "i'll": "i will",
        "don't": "do not",
        "doesn't": "does not",
        "didn't": "did not",
        "haven't": "have not",
        "hasn't": "has not",
        "hadn't": "had not",
        "won't": "will not",
        "wouldn't": "would not",
        "shouldn't": "should not",
        "couldn't": "could not",
        "isn't": "is not",
        "aren’t": "are not",
        "wasn’t": "was not",
        "weren’t": "were not",
        "what's": "what is",
        "where's": "where is",
        "who's": "who is",
    }
    for contraction, fixed_contraction in common_contractions.items():
        if contraction in D:
            D = D.replace(contraction, fixed_contraction)
            print(contraction)
    #Convert Number to word (i know that we don't need it in those documents but any way :) )
    digits = {
        '0': 'zero',
        '1': 'one',
        '2': 'two',
        '3': 'three',
        '4': 'four',
        '5': 'five',
        '6': 'six',
        '7': 'seven',
        '8': 'eight',
        '9': 'nine'
    }

    for digit, word in digits.items():
        if digit in D:
            D = D.replace(digit, word)

    #Tokenization using split
    token_ = D.split(" ")
    token_ = set(token_)
    return token_
```

```python
print('D1 After preprocessing (using split as tokenizer) :',preprocessing_document(D1))
print('D2 After preprocessing (using split as tokenizer):',preprocessing_document(D2))
print('D3 After preprocessing (using split as tokenizer):',preprocessing_document(D3))
```

```python
def tokenization_spacy(D):
    nlp = spacy.load("en_core_web_sm")
    D = D.lower()
    doc = nlp(D)
    token_ = [token.text for token in doc if not token.is_punct]
    token_ = set(token_)
    return token_
```

```python
print('D1 After preprocessing (using spacy as tokenizer) :',tokenization_spacy(D1))
print('D2 After preprocessing (using spacy as tokenizer):',tokenization_spacy(D2))
print('D3 After preprocessing (using spacy as tokenizer):',tokenization_spacy(D3))
```

```python
def text_vectorization(D,vocabulary_vector):
    D = preprocessing_document(D)
    vector = []
    for v in vocabulary_vector:
        vector.append(1 if v in D else 0)
    return vector
```

```python
#considering vocabulary vector in tp guide
vocabulary_vector = ['feel', 'good', 'better', 'thank', 'happened']
print('Vector[D1] :',text_vectorization(D1,vocabulary_vector))
print('Vector[D2] :',text_vectorization(D2,vocabulary_vector))
print('Vector[D3] :',text_vectorization(D3,vocabulary_vector))
```
